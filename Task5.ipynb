```python
data2=pd.read_csv('deliveries.csv')
data2.head()
```

#**Exploring Data**

**Merge the data**

```python
season_data=data1[['id','season','winner']]
complete_data=data2.merge(season_data,how='inner',left_on='match_id',right_on='id')
complete_data.columns
```

**Droping umpire all umpire one by one**

```python
data1 = data1.drop(columns=["umpire2"],axis=1)
data1.head()
```

**Season Winners**

```python
final_matches=data1.drop_duplicates(subset=['season'], keep='last')
final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

**Matches per Season**

```python
plt.figure(figsize = (10,10))
sns.countplot('season',data=data1)
plt.title('No of matches per Season')
plt.show()
```

**No of matches won by team**

```python
plt.figure(figsize=(20,10))
plt.title('Wins per Team',fontsize=20)
sns.countplot(x='winner',data=data1, palette='cool')
plt.xticks(rotation=50)
plt.show()
```

**Match result based on Bat first or bown first**

```python
data1['win_by']=np.where(data1['win_by_runs']>0,'Bat first','Bowl first')
Win=data1.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels,colors=['green','blue'],shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal',fontsize=10)
plt.show()
```

**Number of Matches Won Based on Batting First and Bowling First**

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue='win_by',data=data1,palette='cool')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

**No of wins on based of toss**

```python
Toss=final_matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
plt.figure(figsize = (5,8))
plt.pie(sizes, labels=labels,shadow=True,startangle=90)
plt.title('Toss Result',fontsize=20)
plt.axis('equal',fontsize=10)
plt.show()
```

**Toss winner, Toss decision, Season Winner (Final Match)**

```python
final_matches[['toss_winner','toss_decision','winner']].reset_index(drop=True)
```

**Man of the match in final**

```python
final_matches[['winner','player_of_match']].reset_index(drop=True)
```

**Top Players**

```python
plt.figure(figsize = (18,10))
top_players = data1.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="hsv");
plt.show()
```

**No of Season won by each team**

```python
final_matches['winner'].value_counts()
```

**No of Fours by each teams**

```python
four=complete_data[complete_data['batsman_runs']==4]
four.groupby('batting_team')['batsman_runs'].agg([('Total runs by fours','sum'),('fours','count')]).sort_values('fours',ascending=False)
```

**No of fours by Player**

```python
four_data=complete_data[complete_data['batsman_runs']==4]
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
plt.title("Numbers of fours hit by players ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

**No of fours each Season**

```python
four_data=complete_data[complete_data['batsman_runs']==4]
batsman_four=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index()
ax=batsman_four.iloc[:12,:].plot('season','four',kind='bar',color='brown')
plt.title("Numbers of fours per season",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Season",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

**No of Sixes by team**

```python
sixes=complete_data[complete_data['batsman_runs']==6]
sixes.groupby('batting_team')['batsman_runs'].agg([('Total runs by six','sum'),('sixes','count')]).sort_values('sixes',ascending=False)
```

**No of six by player**

```python
six=complete_data[complete_data['batsman_runs']==6]
batsman_six=six.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='orange')
plt.title("Numbers of sixes hit by players ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of Sixes",fontsize=15)
plt.show()
```

**No of sixes per Season**

```python
six=complete_data[complete_data['batsman_runs']==6]
batsman_six=six.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index()
ax=batsman_six.iloc[:12,:].plot('season','six',kind='bar',color='red')
plt.title("Numbers of Sixes per season",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Season",fontsize=15)
plt.ylabel("No of sixes",fontsize=15)
plt.show()
```

**Top Batsman**

```python
batsman_score=complete_data.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

**Top Wicket Taker**

```python
wicket_data=complete_data.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
